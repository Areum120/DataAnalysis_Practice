```python
weather = pd.read_csv('data2/weather.csv')
weather.iloc[:5, :11]
```

```python
#일별 온도 측정값 피벗한다.
weather_melt = pd.melt(weather,
                      id_vars=['id','year','month','element'],
                      var_name='day',
                      value_name='temp')
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],#위치를 유지할 열 이름
    columns ='element',#피벗할 열 이름
    values='temp'#새로운 열의 데이터가 될 이름
)
weather_tidy
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat.head()
```

```python
billboard = pd.read_csv('data2/billboard.csv')
billboard_long = pd.melt(billboard,
                        id_vars=['year','artist','track','time','date.entered'],
                         var_name = 'week',
                         value_name = 'rating')
billboard_long.shape
billboard_long.head()
```

```python
billboard_long[billboard_long.track =='Loser'].head()
```

```python
# - 중복 데이터를 가지고 있는 열은 따로 모아 새로운 데이터프레임에 저장한다.
billboard_songs = billboard_long[['year','artist','track','time']]
billboard_songs.shape
```

```python
#  drop_duplications 메서드로 데이터프레임의 중목 데이터를 제거한다.
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
# - 중복제거한 데이터프레임에 아이디(id) 열을 추가한다.
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs.head(n=10)
                              
```

```python
# merge 메서드로 노래정보와 주간 순위 데이터 합치기
billboard_ratings = billboard_long.merge(
    billboard_songs,
    on=['year','artist','track','time'])

billboard_ratings.shape
```
