```python
# 분석에 사용할 속성 선택, 연비, 실린더, 엔진출력, 중량
ndf = df[['mpg', 'cylinders','horsepower','weight']]
ndf
```

## 단순회귀분석 ##

```python
# 단순회귀분석은 일대일 대응이므로 변수 선택
# 두 변수 간에 선형관계인 상관관계를 보기 위해 산점도를 그린다
# 연비&중량
ndf.plot(kind='scatter', x = 'weight', y='mpg', c = 'coral', s=10, figsize=(10,5))
plt.show()
plt.close()
# ->중량이 가벼울수록 연비가 높다
```

```python
# 연비&중량 seaborn으로 산점도 그리기 + 회귀선 표시 
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
sns.regplot(x = 'weight', y ='mpg', data=ndf, ax=ax1)#회귀선 표시
sns.regplot(x = 'weight', y ='mpg', data=ndf, ax=ax2, fit_reg=False)#회귀선 미표시
plt.show()
plt.close()
```

```python
# 산점도+히스토그램

sns.jointplot(x='weight', y='mpg', data=ndf)
sns.jointplot(x='weight', y='mpg', kind='reg',data=ndf)#회귀선표시
plt.show()
plt.close()
```

```python
# 데이터프레임 열을 두 개씩 짝 지을 수 있는 모든 경우의 수에 대해 두 변수간의 산점도 그리기
# 자기 자신은 히스토그램
grid_ndf = sns.pairplot(ndf)
plt.show()
plt.close()
```

```python
# 앞의 산점도에서 선형관계를 그린 horsepower,weight를 독립변수 x로 선택
# 두 변수간의 회귀 방정식 찾기
# 훈련 데이터와 검증 데이터로 나눠서 모형 구축
# weight열을 독립 변수x로 선택하고 데이터를 7:3의 비율로 분할

# 1. 훈련용/검증용 데이터로 분할하기

# 변수 선택
X = ndf[['weight']]#독립변수
y = ndf['mpg']#종속 변수

# train data와 test data로 구분(7:3)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)#랜덤 추출값

print('train data 개수 :', len(X_train))
print('test data 개수 :', len(X_test))
```

```python
# 2. 모형 학습 및 검증

# 선형회귀분석 모듈 사용

from sklearn.linear_model import LinearRegression

# 단순회귀분석 모형 객체 생성
Ir = LinearRegression()

# 앞에서 만든 train data를 가지고 모형 학습
Ir. fit(X_train, y_train)

# 학습을 마친 모형에 test data를 적용하여 결정계수(R-제곱) 계산
r_square = Ir.score(X_test, y_test)
r_square

# 결정계수 값이 클수록 모형 예측능력이 좋다.
```

```python
# 위 결정계수와는 별개로 계수 a는 회귀식의 기울기(모형 객체 Ir의 coef_ 속성값)
# 계수 b는 y절편이고 (모형객체 Ir의 intercept_ 속성값)

print('기울기 a:', Ir.coef_)
print('\n')
print('y절편 b:', Ir.intercept_)
```

```python
# 모형의 예측값과 실제 값 비교하기(모형의 오차 확인)

# 모형에 전체 X 데이터를 입력하여 예측한 값 y_hat을 실제 값 y와 비교

y_hat = Ir.predict(X)

plt.figure(figsize=(10, 5))

ax1 = sns.distplot(y, hist=False, label= "y")#파란색
ax2 = sns.distplot(y_hat, hist=False, label ="y_hat", ax = ax1)#주황색
plt.show()
plt.close()

# 실제 값은 왼쪽, 모형은 오른쪽, 독립변수 weight와 종속 변수 mpg 사이에 선형 관계가 있지만 모형의 오차를 줄일 필요
# 앞에서 그린 산점도를 보면 직선보다 곡선에 더 적합-> 비선형 회귀분석을 통해 정확도를 높일 수 있음

```
